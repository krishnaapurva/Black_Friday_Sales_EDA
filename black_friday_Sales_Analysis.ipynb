### **As we can see that we have 2 observation which we need to work on**
##### 1. Need to deal with Null values in Product_Category_2 and Product_Category_3
#####   *We are approaching it by replacing null values with the median of the column*
##### 2. We need to convert the float values into integer values.

```python
pc2_median = df.Product_Category_2.median()
df['Product_Category_2'] = df['Product_Category_2'].fillna(pc2_median).astype(int)
```

```python
pc3_median = df.Product_Category_3.median()
df['Product_Category_3'] = df['Product_Category_3'].fillna(pc3_median).astype(int)
```
